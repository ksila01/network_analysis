### Distribution of Twitter Followers

```python

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['Twitter Followers'], bins=30, kde=True)
plt.title('Distribution of Twitter Followers')
plt.xlabel('Number of Followers')
plt.ylabel('Frequency')
plt.show()
```

Interpretation
* Highly Skewed Distribution:

* The distribution of Twitter followers is highly right-skewed. This indicates that most of the authors have a relatively small number of followers.
* A large majority of authors have a follower count near the lower end of the scale (close to zero).
* Few Accounts with High Followers:

* There are very few accounts with a very high number of followers, which is why the graph extends far to the right.These high-follower accounts are outliers compared to the majority.
* Fruency:

* The highest bar represents the number of authors with very few followers (likely below a few thousand).
* As the number of followers increases, the frequency of such accounts decreases sharply.

Implications
* Influence Concentration:
* The influence on Twitter is likely concentrated among a few accounts with a very high number of followers.
* These accounts may have a significant impact on the spread and reach of the hashtag #MamaMbogaMoment.
* Role of High-Follower Accounts:
* Identifying these high-follower accounts is crucial for understanding the key influencers in the network.
* These accounts can be targeted for further analysis to see their role in spreading the hashtag.

### Network Data Preparation

```python
# Step 1: Prepare Nodes Data
# Combine Authors and Mentioned Authors to ensure all unique nodes are included
authors = cleaned_data[['Author', 'Twitter Followers', 'Account Type']].drop_duplicates()
mentioned_authors = cleaned_data[['Mentioned Authors']].dropna()
```

```python
# Split multiple mentioned authors and clean the data
mentioned_authors = mentioned_authors['Mentioned Authors'].str.replace('@', '').str.split(', ')
mentioned_authors = mentioned_authors.explode().drop_duplicates().reset_index(drop=True)
mentioned_authors = mentioned_authors.to_frame(name='Author')
```

```python
# Merge both authors and mentioned authors
all_authors = pd.merge(authors, mentioned_authors, on='Author', how='outer')
all_authors['Twitter Followers'] = all_authors['Twitter Followers'].fillna(0)
all_authors['Account Type'] = all_authors['Account Type'].fillna('Unknown')
all_authors = all_authors.drop_duplicates().reset_index(drop=True)
all_authors['Id'] = all_authors.index
```

```python
# Save the Nodes file
# Save the Nodes file
nodes_file_path = 'nodes.csv'
all_authors.to_csv(nodes_file_path, index=False)
```
