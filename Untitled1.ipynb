```python
import pandas as pd
```

```python
data = pd.read_csv('MamaMbogaMoment.csv')
```

```python
data.head ()
```

```python
# Extract unique authors as nodes
nodes = data[['Author', 'Twitter Followers', 'Twitter Following']].drop_duplicates()

# Extract edges based on mentions
mentions = data[['Author', 'Mentioned Authors']]
mentions = mentions.dropna()

# Split mentions into individual mentions and create edges
edges = mentions.assign(Mentioned_Authors=mentions['Mentioned Authors'].str.split(',')).explode('Mentioned_Authors')
edges = edges.rename(columns={'Author': 'Source', 'Mentioned_Authors': 'Target'})

# Remove '@' from the mentioned authors
edges['Target'] = edges['Target'].str.strip().str.replace('@', '')

# Save nodes and edges to CSV
nodes.to_csv('nodes.csv', index=False)
edges.to_csv('edges.csv', index=False)
```
